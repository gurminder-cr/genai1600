```python
# Artificial Neural Network
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset= tf.keras.datasets.mnist.load_data()
```

#The MNIST dataset is a foundational machine learning benchmark containing 70,000 grayscale 28x28 pixel images of handwritten digits (0–9). It consists of 60,000 training and 10,000 testing

```python
(x_train, y_train),(x_test, y_test)= dataset
```

```python
x_train_flat =  x_train.reshape(len(x_train),28*28)
```

```python
x_train_flat.shape
```

```python
x_train_flat[0]
```

```python
x_test_flat=  x_test.reshape(len(x_test),28*28)
```

```python
x_test_flat.shape
```

```python
from tensorflow import keras
```

```python
model= keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='softmax')   #[0.45,0.67,0.54,0.78,]
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_flat,y_train,epochs=5)
```

```python
y_pred= model.predict(x_test_flat)
```
