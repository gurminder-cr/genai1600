```python
# pip install scikit-learn
from sklearn.preprocessing import LabelEncoder 
le= LabelEncoder()
```

```python
import pandas as pd
df= pd.DataFrame({'result':['good','bad','good','bad','excellent','bad','excellent','good','bad']})
df
```

```python
df.result.value_counts()
```

```python
# fit, transform, fit_transform
le.fit(df.loc[:,'result'])
```

```python
ans= le.transform(df.loc[:,'result'])
```

```python
df.loc[:,'new']=ans
```

```python
df
```

```python
df.new.value_counts()   
```

```python
encoder= LabelEncoder()
```

```python
encoder.fit_transform(df.loc[:,'result'])
```

```python
import joblib 
```

```python
joblib.dump(le,'encoder.pkl')
```
