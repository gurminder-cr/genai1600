```python
import pandas as pd 
import seaborn as sns 
import plotly.express as px 
```

```python
flights= sns.load_dataset('flights')
```

```python
flights
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
```

```python
flights
```

```python
#pivot table 
d=pd.pivot(data=flights,index='month',columns='year',values='passengers')
```

```python
d
```

```python
sns.heatmap(d,annot=True, fmt="0g",linewidths=2)
```

```python
px.density_heatmap(d)
```

```python
sns.pairplot(data=flights)
```
