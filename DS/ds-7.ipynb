```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

# Bar chart - one axis- Strings value(categorical), 2nd axis - numerical or count 
import matplotlib.pyplot as plt
```

```python
plt.bar(x=fruits,height=sales)
plt.show()
```

```python
plt.barh(y=fruits,width=sales)
plt.show()
```

```python
plt.bar(x=fruits,height=sales, width=0.5, color=['red','green','blue','yellow'])
# plt.grid()
# plt.gray()
plt.xlabel('Fruits')
plt.ylabel('Sales Count')
plt.title('Count of Sales vs Fruits')
plt.show()
```

```python
# pie chart 

country=['India','Pakistan','Bangladesh','Nepal','Spain']
values=[200,180,230,100,120]

plt.pie(x=values, labels=country, autopct="%1.2f%%")
plt.show()
```

```python
import numpy as np
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])
l=[10,50,9,16,18,20,22,21,7,19]

plt.scatter(x=x, y=y, s=l, marker="*")
```
