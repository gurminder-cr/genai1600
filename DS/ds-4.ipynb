```python
# Write a recursive function to compute the sum of digits of a number until a single digit remains.
# Example: 9875 → 9+8+7+5=29 → 2+9=11 → 1+1=2

def funcRe(a):
    sum=0
    for i in a:
        i=int(i)
        sum+=i
    if len(str(sum))==1:
        print(sum)
    else:
        funcRe(str(sum))


funcRe("99913234")        
```
